```python
import time
import random
```

```python
def RandList (start, end, len):
        L=[] # an empty list

        for j in range (len):
                L.append (random.randint(start,end))
                
        return L
```

```python
def bubbleSort(L):
    if length == 0 or length == 1:
        return L
    for i in range(length):
        for j in range(length-i-1):
            if L[j] > L[j+1]:
                temp = L[j]
                L[j] = L[j+1]
                L[j+1] = temp
    return L

def bubbleSortRecursive(L, n):
 
    for i in range(n - 1):
        if L[i] > L[i + 1]:
            L[i], L[i+1] = L[i+1], L[i]
            
    if n - 1 > 1:
        bubbleSortRecursive(L,n - 1)
    
    return L    
```

```python
# bubbleSort
for i in range(10,2000,20): 
        l=RandList(1,10000,i)
        length = len(l)
        start_time = time.time()
        bubbleSort(l)
        end_time = time.time()
        print (i, 'elapsed:', end_time-start_time)	
```

```python
# bubbleSort recursive
for i in range(10,2000,20): 
        l=RandList(1,10000,i)
        n = len(l)
        start_time = time.time()
        bubbleSortRecursive(l,n - 1)
        end_time = time.time()
        print (i, 'elapsed:', end_time-start_time)
```

```python
def InsertionSort(arr):
    if type(arr) is not type([1]):
        return "ERROR INPUT"
    i = 1
    while i < len(arr):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
        i += 1
    return arr
```

```python
# InsertionSort
for i in range(10,2000,20): 
        l=RandList(1,10000,i)
       
        start_time = time.time()
        InsertionSort(l)
        end_time = time.time()
        print (i, 'elapsed:', end_time-start_time)	
```

```python
def FindMinimum (l):
        listlen = len(l)

        minV = l[0]
        minI = 0

        for i in range (1,listlen): 
                if l[i]<minV:
                        minV = l[i]
                        minI = i

        return minI

def FindMinimum (l, left, right):

        minV = l[left]
        minI = left

        for i in range (left,right+1): 
                if l[i]<minV:
                        minV = l[i]
                        minI = i

        return minI
```

```python
def SelectionSort (l):
        listlen=len(l)

        for i in range(listlen): 
                minIndex = FindMinimum (l, i, listlen-1)

                if minIndex!=i:
                        l[i],l[minIndex]=l[minIndex],l[i]


def SelectionSortRecursive (l,left,right):

           if left==right:
               return

           minIndex = FindMinimum (l, left, right)

                
           if minIndex!=left:
                 l[left],l[minIndex]=l[minIndex],l[left]

           SelectionSortRecursive (l, left+1, right)
```

```python
def search(arr, n, x): 
  
    for i in range (0, n): 
        if (arr[i] == x): 
            return i; 
    return -1; 
  
# Find the character D in the array.
arr = [ 'A', 'B', 'C', 'D', 'E' ]; 
x = 'D'; 
n = len(arr); 
result = search(arr, n, x) 
if(result == -1): 
    print("Element is not in the array") 
else: 
    print("The index of the element in the array is", result)
```

```python
def binarySearch (arr, l, r, x): 
    if r >= l: 
        mid = int(l + (r - l)/2)
        if arr[mid] == x: 
            return mid 
        elif arr[mid] > x: 
            return binarySearch(arr, l, mid-1, x) 
        else: 
            return binarySearch(arr, mid+1, r, x) 
    
    else:
        return -1
  
# test
arr = [ 2, 3, 4, 10, 40 ] 
x = 10
  
result = binarySearch(arr, 0, len(arr)-1, x) 
  
if result != -1: 
    print ("The index of the element in the array is%d" % result )
else: 
    print ("Element is not in the array")
```
